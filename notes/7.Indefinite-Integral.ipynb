## 不定积分表

不定积分表是数学分析的基础知识

$$
\begin{aligned}
&\int dx = C\\
&\int 1 dx = x + C\\
&\int x^\mu dx = \frac{1}{\mu+1}x^{\mu+1} + c(\mu \neq -1)\\
&\int x^{-1}dx = \int \frac{dx}{x}=ln\left|x\right|+C\\
&\int \frac{dx}{x-a} = ln\left|x-a\right|+c\\
&\int e^xdx = e^x +C\\
&\int a^xdx=\frac{a^x}{ln a}+C\left(a\gt 0, a\neq 0\right)\\
&\int cos x dx=sin x +C\\
&\int sin x dx=- cos x + C \\
&\int \frac{dx}{cos^2 x} = tan x + C\\
&\int \frac{dx}{sin^2 x} = - cot x +C\\
&\int \frac{dx}{1+x^2} = arctan x +C\\
&\int \frac{dx}{\sqrt{1-x^2}} = arcsin x +C\\
&\int ch x dx = sh x + C\\
&\int sh x dx = ch x + C\\
&\int \frac{dx}{\sqrt{x^2+a^2}} = ln\left|x+\sqrt{x^2+a^2}\right| + C
\end{aligned}
$$



## 不定积分的线性特性

如果 $F(x)$ 和 $G(x)$ 分别是函数 $f(x)$ 和 $g(x)$ 原函数， $\lambda$ 是一个实数，那么 $F(x) + G(x)$ 是函数 $f(x)+g(x)$ 的原函数，换句话说，我们有以下运算法则：

$$
\begin{aligned}
\int\left(f(x)+g(x)\right)dx&=\int f(x) + \int g(x) dx;\\
\int\left(\lambda f(x)\right)dx&=\lambda\int f(x)dx
\end{aligned}
$$

## 换元积分法

如果 

$$
dG(\mu) = g(\mu)d\mu
$$

那么把 $\mu$ 换成可微函数 $\mu=\mu(v)$ 仍有

$$
dG(\mu(v)) = g(\mu(v))d\mu(v)
$$

这就是说，从

$$
\int g(x)d\mu = G(\mu)+C
$$

可以得到

$$
\int g(\mu(v))d\mu(v) = G(u(v)) +C
$$

### 第一换元法

记录一下典型的例题。

$$
\int g(ax +b) dx = \frac{1}{a}\int g(ax+b)d(ax+b)
$$

$$
\int g(x^k)d^{k-1}dx=\frac{1}{k}\int g(x^k)d(x^k)
$$

$$
\int \frac{g(ln x)}{x} dx = \int g(ln x)d(ln x)
$$

$$
\int \frac{e^x}{1+e^{2x}}dx = \int \frac{d e^x}{1+ (e^x)^2} = \arctan e^x +C
$$

$$
\int sin^2 xdx = \int \frac{1 - cos^2x}{2}dx = \frac{1}{2}x - \frac{1}{4}sin2x+C
$$

$$
\int cos^2 x dx = \int \frac{1+cos2x}{2}dx=\frac{1}{2}x+\frac{1}{4}sin2x+C
$$

$$
\int \tan x dx = - \int \frac{d(\cos x)}{\cos x } = - ln \left| \cos x \right| +C
$$

$$
\int \cot x dx = \int \frac{d(\sin x)}{\sin x} = ln \left|\sin x\right| + C
$$

$$
\int \cos^3 x dx= \sin x - \frac{1}{3}\sin^3 x +C
$$

$$
\int \sin^3 x dx = - \cos x + \frac{1}{3}\cos^3 x + C
$$

$$
\int \frac{dx}{\cos x}=ln\left|\sec x + \tan x\right|+C
$$

$$
\int \frac{dx}{\sqrt{a^2-x^2}}=\int \frac{d\frac{x}{a}}{\sqrt{1 - \frac{x}{a}^2}} = \arcsin \frac{x}{a}+C
$$

$$
\int \frac{dx}{a^2+x^2} = \frac{1}{a}\int\frac{d\frac{x}{a}}{1+\left(\frac{x}{a}\right)^2}=\frac{1}{a}\arctan \frac{x}{a}+C
$$

$$
\int\frac{dx}{x^2+px+q} =
\begin{cases}
\frac{1}{\alpha-\beta}\ln\left|\frac{x-\alpha}{x-\beta}\right| + C & (x-\alpha)(x-\beta)=0;\alpha \neq \beta\\
-\frac{1}{x-\gamma}+C & x^2 + px + q = (x -\gamma)^2\\
\frac{1}{\sqrt{q-\frac{p^2}{4}}}\arctan \frac{x + \frac{p}{2}}{\sqrt{q-\frac{p^2}{4}}}+C &x^2 + px + q = (x-(\lambda\pm i \mu))^2
\end{cases}
$$


### 第二换元法

对于$\int f(x) dx$，如果 $x=\varphi(t)$ 且 $\varphi(t)$ 在区间 $J$ 上严格单调、连续、可导——即在区间$J$上$\varphi^{-1}$有意义，需要注意这是一个充分不必要的定义，此条件比函数逆存在更强——并满足 $\varphi'(t)\neq0$，并求得

$$
\int f(\varphi(t)) d \varphi(t) = G(t) + C
$$

那么令 $t=\varphi^{-1}(x)$ 就得到

$$
\int f(x)dx=G(\varphi^{-1}(x)) + C
$$

#### 例 $\int \sqrt{a^2-x^2}dx$

令 $x = a\sin t (-\frac{\pi}{2}\le t \le \frac{\pi}{2})$

$$
\int \sqrt{a^2-x^2}dx=\frac{1}{2}\left(a^2\arcsin\frac{x}{a}+x\sqrt{a^2-x^2}\right)+C
$$

#### 例 $\int \frac{dx}{\left(x^2+a^2\right)^2}$

$$
\int \frac{dx}{\left(x^2+a^2\right)^2} = \frac{1}{2a^3}\arctan\frac{x}{a}+\frac{1}{2a^2}\frac{x}{x^2+a^2}+C
$$

#### 例 $\int\frac{dx}{\sqrt{x^2+a^2}}$

$$
\int\frac{dx}{\sqrt{x^2+a^2}}=\ln\left|x+\sqrt{x^2+a^2}\right|+C
$$

#### 例 $\int\frac{dx}{\sqrt{x^2-a^2}}$

这里 $\left|x\right| \gt a \gt 0$

有

$$
\int\frac{dx}{\sqrt{x^2-a^2}}=\ln\left|x + \sqrt{x^2-a^2}\right|+C
$$

### 分部积分法

将公式
$$
d(\mu(x)v(x)) = v(x)d\mu(x) + \mu(x)dv(x)
$$

改写为
$$
\mu(x)dv(x) = d(\mu(x)v(x))-v(x)du(x)
$$

由此得到

$$
\int\mu(x)dv(x) = \mu(x)v(x) - \int v(x)d\mu(x)
$$

由此得到以下示例：

$$
\int x \cos x dx = x\sin x + \cos x + C
$$

$$
\int x \sin x dx = - x \cos x + \sin x + C
$$

$$
\int x e^x dx = (x-1)e^x + C
$$

$$
\int x^k\ln x dx = 
\begin{cases}
\frac{1}{k+1}x^{k+1}\ln x - \frac{1}{\left(k+1\right)^2}x^{k+1} +C&x\neq-1\\
\frac{1}{2}(\ln x)^2+C& x=-1
\end{cases}
$$

$$
\int x \arctan x dx = \frac{1}{2}(x^2+1)\arctan x - \frac{1}{2}x +C
$$

$$
\int x^2\cos x dx=x^2\sin x+2x \cos x - 2 \sin x +C
$$

$$
\int \sqrt{x^2 - a^2} dx = \frac{x}{2}\sqrt{x^2-a^2} - \frac{a^2}{2}\ln\left|x+\sqrt{x^2-a^2}\right|+C
$$

$$
\int \sqrt{x^2 + a^2} dx = \frac{x}{2}\sqrt{x^2+a^2} - \frac{a^2}{2}\ln\left|x+\sqrt{x^2+a^2}\right|+C
$$

$$
\int e^{ax}\cos bx dx = e^{ax}\frac{a\cos bx+b\sin bx}{a^2+b^2}+C
$$

$$
\int e^{ax}\sin bx dx = e^{ax}\frac{a\sin bx-b\cos bx}{a^2+b^2}+C
$$

求
$$
\begin{aligned}
J_n &= \int \frac{dx}{\left(x^2+a^2\right)^n}\\
&= \frac{x}{x^2+a^2} - \int xd\frac{1}{\left(x^2+a^2\right)^n}\\
&= \frac{x}{x^2+a^2} + 2n\int \frac{x^2}{\left(x^2+a^2\right)^{n+1}}dx\\
&= \frac{x}{x^2+a^2} + 2n\int \frac{x^2+a^2-a^2}{\left(x^2+a^2\right)^{n+1}}dx\\
&= \frac{x}{\left(x^2+a^2\right)^n}+2nJ_n-2na^2J_{n+1}
\end{aligned}
$$
得到
$$
J_{n+1}=\frac{1}{2na^2}\frac{x}{\left(x^2+a^2\right)^n}+\frac{2n-1}{2na^2}J_n
$$
和
$$
J_1=\int\frac{dx}{x^2+d^2} = \frac{1}{a}\arctan\frac{x}{a}+C
$$
